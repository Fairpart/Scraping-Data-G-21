```python
# import library
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
data = requests.get('https://pantip.com/tag/%E0%B8%84%E0%B8%AD%E0%B8%99%E0%B9%82%E0%B8%94%E0%B8%A1%E0%B8%B4%E0%B9%80%E0%B8%99%E0%B8%B5%E0%B8%A2%E0%B8%A1%E0%B9%83%E0%B8%81%E0%B8%A5%E0%B9%89%E0%B8%A3%E0%B8%96%E0%B9%84%E0%B8%9F%E0%B8%9F%E0%B9%89%E0%B8%B2')
soup = BeautifulSoup(data.text)
```

```python
pantipDict = {}

listing = soup.find_all('li',{'class':'pt-list-item'})
for num in range(len(listing)):
    tempDict={}
    topicText = listing[num].find('h2').get_text()
    tempDict['Topic'] = topicText
    
    tagList = listing[num].find('div',{'class':'pt-list-item__tag'})
    tempList = []
    for tag in tagList.find_all('a'):  
        tempList.append(tag.get_text())
        tempList_str = str(tempList)
        tempList_str = tempList_str.strip("[]")
        tempDict['Tag'] = tempList_str
        pantipDict[num] = tempDict 
        
    dateText = listing[num].find('div',{'class':'pt-list-item__info'})
    for dt in dateText.find_all('span'):  
        tempDict['Date Posted'] = dt.get_text() 
        pantipDict[num] = tempDict 
        
    NoComtext = listing[num].find('div',{'class':'pt-list-item__stats'})
    for n in NoComtext.find_all('span',{'class':'pt-li_stats-comment'}):
        tempDict['No. of Comments'] = n.get_text()[7:] 
        pantipDict[num] = tempDict 

pantipDict
```

```python
table = pd.DataFrame.from_dict(pantipDict).transpose()
table
```

```python
table['Topic'] = table['Topic'].astype(str)
table['Tag'] = table['Tag'].astype(str)
table['Date Posted'] = table['Date Posted'].astype(str)
table['No. of Comments'] = table['No. of Comments'].astype(int)
```
